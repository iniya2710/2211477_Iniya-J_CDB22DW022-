```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

# Joint plot

```python
sns.jointplot(x='fare',y='age',data=df)
```

# Distribution plot

```python
sns.distplot(df['fare'],kde=False,color='red',bins=30)
```

# Box plot

```python
sns.boxplot(x='class',y='age',data=df,palette='rainbow')
```

# Swarm plot 

```python
sns.swarmplot(x='class',y='age',data=df,palette='Set2')
```

# Count plot

```python
sns.countplot(x='sex',data=df)
```

# Heat map

```python
tc = df.corr()
sns.heatmap(tc,cmap='coolwarm')
```
